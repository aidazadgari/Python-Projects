Exercise 1: Define a function called Recipe which takes two items: one
is name of recipe and one is ingredient. It has two print statements:
one which prints name of recipe and one which prints the ingredient. 
1- Spaghetti Bolognese : Pasta and mince
2- Tomato Soup : Tomatoes and water
3- Steak Tartare : Raw Steak with Egg

```python
def Recipe(NameOfRecipe, Ingredient):

    print(f'Recipe:{NameOfRecipe}')
    print(f'Ingredient: {Ingredient}')
    print("="*10)
```

```python
Recipe('Spaghetti Bolognese','Pasta and mince')
Recipe('Tomato Soup','Tomatoes and water')
Recipe('Steak Tartare','Raw Steak with Egg')
```

```python
def Recipe(NameOfRecipe, Ingredient):

    print('Recipe is %s' % (NameOfRecipe))
    print('Ingredient is %s' % (Ingredient))
    print("="*10)
```

```python
Recipe('Spaghetti Bolognese', 'Pasta and mince')
```

Exercise: Define a function called WashLaundry which takes a (single)
item (e.g. shirt) and prints ‘a message’
that says “Washing item” e.g. Washing shirt, Washing Blouse etc.

```python
def WashLaundry(item):
    print(f'Washing {item}')
```

```python
WashLaundry('shirt')
WashLaundry('Blouse')
```
