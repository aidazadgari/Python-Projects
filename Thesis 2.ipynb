```python
import statsmodels.api as sm
from linearmodels.panel import PooledOLS

exog_vars = ['CR', 'LR', 'SIZE']
exog = sm.add_constant(data[exog_vars])
mod = PooledOLS(data.ROA, exog)
pooled_res = mod.fit()
print(pooled_res)
```

```python
# fixed effect
```

```python
from linearmodels.panel import PanelOLS
mod = PanelOLS.from_formula('ROA ~  CR + LR + SIZE + EntityEffects + TimeEffects',data=data)
fe_res = mod.fit(cov_type='clustered', cluster_entity=True)
print(fe_res)
```

```python
fe_res.estimated_effects
fe_res_effects = fe_res.estimated_effects.unstack(level=0).values[0]
print(fe_res_effects)
```

```python
from linearmodels.panel import compare
print(compare({"FE": fe_res, "RE": re_res, "Pooled": pooled_res}))
```

```python
exog_vars = ['CR', 'LR', 'SIZE']
exog = sm.add_constant(data[exog_vars])
mod = PanelOLS(data.ROA, exog, entity_effects=True, time_effects=True)
result = mod.fit(cov_type='clustered', cluster_entity=True)
print(result)
```
